## Lab Task 1: Image Basics and Preprocessing using OpenCV

### Objective
Learn how to load an image, understand image shape, convert color spaces, resize, crop, flip, blur, and adjust brightness/contrast.

### Required Output
- Original image
- RGB image
- Grayscale image
- Resized image
- Cropped image
- Flipped image
- Blurred image
- Brightness/contrast adjusted image

```python
# Install/import required libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
import os

print("OpenCV version:", cv2.__version__)
```

## Step 1: Upload an Image
Upload any `.jpg`, `.jpeg`, or `.png` image.

```python
uploaded = files.upload()

image_path = list(uploaded.keys())[0]
print("Uploaded image:", image_path)
```

## Step 2: Read and Display Original Image
OpenCV reads images in **BGR** format by default.

```python
img_bgr = cv2.imread(image_path)

if img_bgr is None:
    raise ValueError("Image could not be loaded. Please upload a valid image file.")

print("Image shape:", img_bgr.shape)
print("Height:", img_bgr.shape[0])
print("Width:", img_bgr.shape[1])
print("Channels:", img_bgr.shape[2] if len(img_bgr.shape) == 3 else 1)

plt.figure(figsize=(6,5))
plt.imshow(img_bgr)
plt.title("Original Image Displayed Directly by OpenCV BGR")
plt.axis("off")
plt.show()
```

## Step 3: Convert BGR to RGB
Matplotlib expects images in **RGB**, so we convert BGR to RGB.

```python
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(6,5))
plt.imshow(img_rgb)
plt.title("RGB Image")
plt.axis("off")
plt.show()
```

## Step 4: Convert Image to Grayscale

```python
gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)

plt.figure(figsize=(6,5))
plt.imshow(gray, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")
plt.show()

print("Grayscale shape:", gray.shape)
```

## Step 5: Resize Image

```python
resized = cv2.resize(img_rgb, (300, 300))

plt.figure(figsize=(5,5))
plt.imshow(resized)
plt.title("Resized Image: 300 x 300")
plt.axis("off")
plt.show()

print("Resized shape:", resized.shape)
```

## Step 6: Crop Region of Interest
Cropping syntax: `image[y1:y2, x1:x2]`.

```python
h, w = img_rgb.shape[:2]

# Crop center region automatically
x1, x2 = w//4, 3*w//4
y1, y2 = h//4, 3*h//4

cropped = img_rgb[y1:y2, x1:x2]

plt.figure(figsize=(5,5))
plt.imshow(cropped)
plt.title("Cropped Center Region")
plt.axis("off")
plt.show()

print("Cropped shape:", cropped.shape)
```

## Step 7: Flip Image

```python
flipped_horizontal = cv2.flip(img_rgb, 1)
flipped_vertical = cv2.flip(img_rgb, 0)

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.imshow(flipped_horizontal)
plt.title("Horizontal Flip")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(flipped_vertical)
plt.title("Vertical Flip")
plt.axis("off")

plt.show()
```

## Step 8: Apply Gaussian Blur

```python
blurred = cv2.GaussianBlur(img_rgb, (9, 9), 0)

plt.figure(figsize=(6,5))
plt.imshow(blurred)
plt.title("Gaussian Blur")
plt.axis("off")
plt.show()
```

## Step 9: Brightness and Contrast Adjustment
Formula: `new_image = alpha * image + beta`

- `alpha` controls contrast
- `beta` controls brightness

```python
alpha = 1.4  # contrast
beta = 40    # brightness

bright_contrast = cv2.convertScaleAbs(img_rgb, alpha=alpha, beta=beta)

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.imshow(img_rgb)
plt.title("Original")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(bright_contrast)
plt.title("Brightness and Contrast Adjusted")
plt.axis("off")

plt.show()
```

## Step 10: Save Processed Images

```python
# Convert RGB images back to BGR before saving with OpenCV
cv2.imwrite("grayscale_image.jpg", gray)
cv2.imwrite("resized_image.jpg", cv2.cvtColor(resized, cv2.COLOR_RGB2BGR))
cv2.imwrite("cropped_image.jpg", cv2.cvtColor(cropped, cv2.COLOR_RGB2BGR))
cv2.imwrite("flipped_image.jpg", cv2.cvtColor(flipped_horizontal, cv2.COLOR_RGB2BGR))
cv2.imwrite("blurred_image.jpg", cv2.cvtColor(blurred, cv2.COLOR_RGB2BGR))
cv2.imwrite("brightness_contrast_image.jpg", cv2.cvtColor(bright_contrast, cv2.COLOR_RGB2BGR))

print("Processed images saved successfully.")
```
